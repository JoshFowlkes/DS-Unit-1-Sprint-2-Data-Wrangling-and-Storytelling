## Plot visualization

```python
now = merged[merged['year'] == 2018]

sns.relplot(x='income', y='lifespan', hue='region', size='population', data=now);


```

## Analyze outliers

```python
qatar_now = now[(now.income > 80000) & (now.country == 'Qatar')].sort_values(by='income')

sns.relplot(x='income', y='lifespan', hue='region', size='population', data=now);
plt.text(x=qatar_now.income-5000, y=qatar_now.lifespan+1, s='Qatar')

plt.show()
```

## Plot multiple years

```python
years = [1818, 1918, 2018]

centuries = merged[merged.year.isin(years)] # this read as "if merged in year is in years" 
                                            #keep it
sns.relplot(x='income', y='lifespan', hue='region', size='population',
            col='year',data=centuries); # col is what made three of them
plt.xscale('log');


#now this tells a story 
```

```python
years = [1818, 1918, 2018]

centuries = merged[merged.year.isin(years)] # this read as "if merged in year is in years" 
                                            #keep it
sns.relplot(x='income', y='lifespan', hue='region', size='population',
            col='year',data=centuries, kind='line'); # added line for 'kind'
plt.xscale('log');

# lol what
```

## Point out a story

```python
new_years = [1918, 1938, 1958, 1978, 1998, 2019] # added more years since WW1

scores = merged[merged.year.isin(new_years)]

sns.relplot(x='income', y='lifespan', hue='region', size='population',
            col='year',data=scores); 
# plt.xscale('log');   # this was messing it up for some reason

# this tells a nice story, cool beans 
```

```python
# can use a loop instead and make a plot for each so that it doesn't get horizonatlly expansive
# like it did here above, 

for year in years: 
  sns.relplot(x='income', y='lifespan', hue='region', size='population',
            data=merged[merged.year==year])
  
  plt.xscale('log')
  plt.xlim((150, 150000))
  plt.ylim(10,90)
  plt.title(year)
  #plt.axhline()
```
